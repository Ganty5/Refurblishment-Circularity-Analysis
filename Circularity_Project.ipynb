```python
# CONDITION BREAKDOWN -  REFURBISHMENT POTENTIAL ANALYSIS
# detecting the devices that are viable for refurbishment
```

```python
print(Rebu['Condition.Score'].dtype)
```

```python
Rebu['Condition.Score'] = pd.to_numeric(Rebu['Condition.Score'], errors='coerce')
```

```python
#using a conditional statement, to determine the conditions for these items to be refurblished.


def categorize_condition(score):
    if score >= 4:
        return 'High'
    elif score >= 2:
        return 'Medium'
    elif score > 0:
        return 'Low'
    else:
        return 'Discard'

Rebu['Condition.Category'] = Rebu['Condition.Score'].apply(categorize_condition)
print(Rebu['Condition.Category'].value_counts())
```

```python
#The analysis above simply categorized the devices into four groups "High, Medium, Low, and Discard",  a lot of the devices fall in the 
#category of discard meaning they are not in good shape to be refurblished or they are very poor in condition.
#This distribution suggests that most of the stock requires significant attention to evaluate for recycling or refurbishing.
#Efforts should focus on identifying efficient ways to process the large number of discarded devices to maximize value recovery.
```

```python
#This heatmap shows the correlation between three variables specifically "Condition Score, RAM.GB and HDD.MB". 
#Using the result, we can measure how closely related two variables are on a scale of -1 to 1.
```

```python
# At 1, it means a Perfect positive correlation such that when one increases, the other also increases.
#At 0, there is no correlation, no relationship at all between the two variables.
# At -1, there is Perfect negative correlation when one increases, the other decreases.


#From the result, we can see that there is a strong correlation between the device condition score and the RAM size,
#simply implying tat devices with larger size of RAM tend to have higher condition scores, could be as a result of its newer state.
#There's a moderately strong positive correlation between the device's condition score and the storage capacity (HDD)
#There is a moderate positive correlation between RAM size and storage capacity (HDD)
```

```python
#The Condition.Score measures the device health or quality and this is closely tied to its RAM and HDD capacity.
# Devices with better specifications are likely in better condition.With this, it is recommended that with this insights,
#Priority should be placed on devices for sales based on their RAM and HDD as it strongly influence the device condition.
```

```python
#ENVIRONMENTAL IMPACT ANALYSIS

#Understanding how refublishing these devices contribute to sustainability
```

```python
#Calculating the percentage of devices in High, Medium and Low condition to determine E-waste Reduction potentials

total_devices = len(Rebu)
refurbished_potential = Rebu[Rebu['Condition.Category'] != 'Discard'].shape[0]
waste_diversion_rate = (refurbished_potential / total_devices) * 100
print(f"Waste Diversion Rate: {waste_diversion_rate:.2f}%")
```

```python

#Estimated CO reduction bu comparing it to the manufacturing of a new device, Assume refurbishing a device saves X kg of CO

co2_saved_per_device = 50  # Example value in kg
total_co2_saved = refurbished_potential * co2_saved_per_device
print(f"Estimated CO Savings: {total_co2_saved:.2f} kg")
```
